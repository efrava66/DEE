```python
#Semana 6 (Funções: Corrente de carga e Corrente Capacitor)
from re import S
import numpy as np
import cmath as cmath
import math as math
  
# Definindo a potencia da carga por fase 
potencia_carga_fase_a_modulo = 1_000_000
potencia_carga_fase_b_modulo = 800_000
potencia_carga_fase_c_modulo = 1_100_000

fp_carga_fase_a = 0.90
fp_carga_fase_b = 0.95
fp_carga_fase_c = 0.85

# Definindo os parametros ZIP
z_percentual = 0.15
i_percentual = 0.25
p_percentual = 0.60

# Definindo a tensão nominal de fase
V_nom_fase = 12_470/(math.sqrt(3))

#######  EXERCICIO PARTE 2  ##########
# Definindo a tensão nominal (parte 2) - comente (Ctrl+/) se for usar a parte 3
# V_carga_fase_a = V_nom_fase
# V_carga_fase_b = V_nom_fase
# V_carga_fase_c = V_nom_fase

# angulo_V_carga_fase_a = 0
# angulo_V_carga_fase_b = -120
# angulo_V_carga_fase_c = 120
######################################

####### EXERCICIO PARTE 3 ############
# Definindo a tensão não nominal (parte 3)  - comente (Ctrl+/) se for usar a parte 2
V_carga_fase_a = 6_851
V_carga_fase_b = 6_973
V_carga_fase_c = 6_886

angulo_V_carga_fase_a = -1.9
angulo_V_carga_fase_b = -122.1
angulo_V_carga_fase_c = 117.5
#######################################

# Listas com os valores de tensão, potencia
V_carga = np.array ([[V_carga_fase_a, angulo_V_carga_fase_a*np.pi/180], [V_carga_fase_b, angulo_V_carga_fase_b*np.pi/180], [V_carga_fase_c, angulo_V_carga_fase_c*np.pi/180]])
S_nom_carga = np.array ([[potencia_carga_fase_a_modulo, fp_carga_fase_a], [potencia_carga_fase_b_modulo, fp_carga_fase_b], [potencia_carga_fase_c_modulo, fp_carga_fase_c]])
ZIP = np.array ([z_percentual, i_percentual, p_percentual])

# Calculo da potencia complexa para indutivo
Sa = complex(S_nom_carga[0,0]*S_nom_carga[0,1], S_nom_carga[0,0]*np.sin(np.arccos(S_nom_carga[0,1]))) # # Se for capacitivo +, indutivo -
Sb = complex(S_nom_carga[1,0]*S_nom_carga[1,1], S_nom_carga[1,0]*np.sin(np.arccos(S_nom_carga[1,1])))
Sc = complex(S_nom_carga[2,0]*S_nom_carga[2,1], S_nom_carga[2,0]*np.sin(np.arccos(S_nom_carga[2,1])))

S_nom_carga_rec = np.array([Sa, Sb, Sc])

def CorrenteCargaY(V_nom_fase,V_carga,S_nom_carga_rec,ZIP):
       
    # Vetor de tensoes atuais na carga
    Van = cmath.rect(V_carga[0,0], V_carga[0,1])
    Vbn = cmath.rect(V_carga[1,0], V_carga[1,1])
    Vcn = cmath.rect(V_carga[2,0], V_carga[2,1])
    Vabcn = np.array ([Van, Vbn, Vcn])

    # Vetor de tensoes nominais na carga
    Van_nom = cmath.rect(V_nom_fase, 0)
    Vbn_nom = cmath.rect(V_nom_fase, -2*np.pi/3)
    Vcn_nom = cmath.rect(V_nom_fase, 2*np.pi/3)

    Sa = S_nom_carga_rec[0]
    Sb = S_nom_carga_rec[1]
    Sc = S_nom_carga_rec[2]

    # Cargas em Y – Impedância constante:
    Za = (abs(Van_nom)**2)/np.conj(Sa)
    Zb = (abs(Vbn_nom)**2)/np.conj(Sb)
    Zc = (abs(Vcn_nom)**2)/np.conj(Sc)

    ILZa = Van/Za
    ILZb = Vbn/Zb
    ILZc = Vcn/Zc

    ILZ = np.array([ILZa, ILZb, ILZc])
    print("Corrente na Carga (A) usando Impedância constante: ")
    print(ILZ)
    print("\n")

    # Cargas em Y – Corrente constante:
    ILa_abs = abs(Sa)/abs(Van_nom)
    ILb_abs = abs(Sb)/abs(Vbn_nom)
    ILc_abs = abs(Sc)/abs(Vcn_nom)

    ILIa = cmath.rect(ILa_abs, np.angle(Van) - cmath.phase(Sa)) # Se for capacitivo -, indutivo +
    ILIb = cmath.rect(ILb_abs, np.angle(Vbn) - cmath.phase(Sb))
    ILIc = cmath.rect(ILc_abs, np.angle(Vcn) - cmath.phase(Sc))

    ILI = np.array([ILIa, ILIb, ILIc])
    print("Corrente na Carga (A) usando Corrente constante: ")
    print(ILI)
    print("\n")

    # Cargas em Y – Potência constante: 
    ILa = np.conj(Sa/Van)
    ILb = np.conj(Sb/Vbn)
    ILc = np.conj(Sc/Vcn)

    
    ILP = np.array([ILa, ILb, ILc])
    print("Corrente na Carga (A) usando Potência constante: ")
    print(ILP)
    print("\n")
           
    # Cargas em Y – ZIP:
    ILZIPa = ZIP[0]*ILZ[0] + ZIP[1]*ILI[0] + ZIP[2]*ILP[0] 
    ILZIPb = ZIP[0]*ILZ[1] + ZIP[1]*ILI[1] + ZIP[2]*ILP[1] 
    ILZIPc = ZIP[0]*ILZ[2] + ZIP[1]*ILI[2] + ZIP[2]*ILP[2] 


    ILZIP = np.array([ILZIPa, ILZIPb, ILZIPc])
    print("Corrente na Carga (A) usando ZIP: ")
    print(ILZIP)
    
    return ILZ, ILI , ILP, ILZIP


Q_nom_Cap_fase = 50

def CorrenteCapacitorY(V_nom_fase,V_carga,Q_nom_Cap_fase):
   
    # Calculo das tensoes atuais de fase
    Van = cmath.rect(V_carga[0,0], V_carga[0,1])
    Vbn = cmath.rect(V_carga[1,0], V_carga[1,1])
    Vcn = cmath.rect(V_carga[2,0], V_carga[2,1])

    # Calculo da susceptancia constante
    B = Q_nom_Cap_fase/(1000*(abs(V_nom_fase/1000))**2)

    # Calculo das correntes de fase no capacitor 
    ICa = complex (0, B*Van)
    ICb = complex (0, B*Vbn)
    ICc = complex (0, B*Vcn)

    ICAP = np.array([ICa, ICb, ICc])
    print("\n")
    print("Corrente nos Capacitores (A):")
    print(ICAP)
    print("\n")
    
    return ICAP

 
ILZIP = CorrenteCargaY(V_nom_fase,V_carga,S_nom_carga_rec,ZIP)[3]
ICAP = CorrenteCapacitorY(V_nom_fase,V_carga,Q_nom_Cap_fase)
total = ICAP + ILZIP

print("Nova Corrente de Linha (A): ")
print(total)
print("\n")
```
